```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Gathering Data

```python
link = lambda x : f"Bike-sharing-dataset\{x}.csv"
```

## membaca dataset

```python
day_df = pd.read_csv(link("day"))
day_df.head()
```

```python
hour_df = pd.read_csv(link("hour"))
hour_df.head()
```

## merge data frame

```python
bike_df = hour_df.merge(day_df, on='dteday', how='inner', suffixes=('_hour', '_day'))
bike_df.head()
```
